# SQL - Select mit Pandas 

```python
pd.read_sql("""
SELECT Name AS `Anzahl der Länder, die mit D beginnen`
FROM cia
WHERE Name LIKE "D%"
ORDER BY Name
""", con)
```

## Group By

```python
pd.read_sql("""
SELECT Region, COUNT(*) AS `Anzahl Länder`
FROM cia
GROUP BY region
""", con)
```

## Group By Having

```python
pd.read_sql("""
SELECT Region, COUNT(*) AS `Anzahl Länder`
FROM cia
GROUP BY region
HAVING `Anzahl Länder` < 10
""", con)
```

# SELECT FROM WHERE GROUP BY HAVING ORDER BY LIMIT
an einem Beispiel

```python
pd.read_sql("""
SELECT Region, COUNT(*) AS `Anzahl Länder`
FROM cia
WHERE Region IS NOT "Welt"
GROUP BY region
HAVING `Anzahl Länder` BETWEEN 10 AND 20
ORDER BY `Anzahl Länder`
LIMIT 2
""", con)
```
